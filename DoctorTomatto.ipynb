```python
#...... inicio librerias para deteccion ..........
from os import listdir
from xml.etree import ElementTree
from numpy import zeros
from numpy import asarray
from numpy import expand_dims
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
from mrcnn.config import Config
from mrcnn.model import MaskRCNN
from mrcnn.model import mold_image
from mrcnn.utils import Dataset
from matplotlib.pyplot import figure
from skimage.io import imread

#....... fin librerias para deteccion ...............

#...... inicio librerias para clasificacion ..........
from keras.applications.imagenet_utils import preprocess_input, decode_predictions
from keras.models import load_model
import cv2
import numpy as np
##import matplotlib.pyplot as plt
%matplotlib inline

#....... fin librerias para clasificacion ...............

#....... inicio librerias para servidor web y comandos os ...............
import os
from os import remove
from flask import Flask, render_template, request,redirect,session,url_for,send_from_directory
from werkzeug.utils import secure_filename

#....... fin librerias para servidor web y comandos os ...............


probabilidad = -1.00  # si la probabilidad es menor a la del umbral definido, se queda con el valor inicial negativo
etiquetas = ("HojaTomate", "NoHojaTomate") # ... clases definidas para deteccion ...
etiqueta=''     # guarda clase detectada ...

app = Flask(__name__) 
# Carpeta de subida de fotos
app.config['UPLOAD_FOLDER'] = './static/upload'

@app.route('/')
def index():
    print('acceso desde ip:',request.remote_addr)  # ... captura Ip de acceso a la aplicacion ...
    return render_template('index.html')
    
@app.route('/objetivo')
def objetivo():
    return render_template('objetivo.html')

@app.route('/subirFoto')
def subirFoto():
    return render_template('subirFoto.html')

@app.route("/fotosSubidas", methods=['POST'])
def uploader(): 
    ALLOWED_EXTENSIONS = set(['jpg', 'jpeg', 'JPG', 'JPEG'])  #...extensiones permitidas ...
    IMAGE_WIDTH,IMAGE_HEIGHT = 800,600         # ... ancho imagen altura imagen ..
    
    def allowed_file(filename):
        return '.' in filename and \
           filename.rsplit('.', 1)[1] in ALLOWED_EXTENSIONS
    
    # ... recupera los nombres de los archivos de la carpeta de archivos subidos ..
    for nombre_archivo in os.listdir('./static/upload/'):
        print('archivo: ',nombre_archivo,len(nombre_archivo))
        remove("./static/upload/"+nombre_archivo)       # ... borrra archivo
        
    for nombre_archivo in os.listdir('./static/output/'):
        print('archivo: ',nombre_archivo,len(nombre_archivo))
        remove("./static/output/"+nombre_archivo)       # ... borrra archivo    

    if request.method == 'POST':
        # obtenemos el nombre del archivo del input "archivo"
        f = request.files['archivo']
        if f.filename=='':       #... controla que no suba foto sin haberla seleccionado ...
            mensaje='<<< ... ERROR ... >>> No puede Subir Foto sin haberla seleccionado ...'
            return render_template('mensaje.html', mensaje=mensaje)

        if f and allowed_file(f.filename):    # ... si se selecciono archivo y tiene formato .JPG o .JPEG 
            filename = secure_filename(f.filename)
            # Guardamos el archivo en el directorio "./static/upload"
            f.save(os.path.join(app.config['UPLOAD_FOLDER'], filename))
            
            # Retornamos una respuesta satisfactoria
            return redirect('/resultado')
        else:
            mensaje='<<< ... ERROR ... >>> Archivo no tiene extensin .JPG  .JPEG, suba otra imagen con el formato requerido.'
            return render_template('mensaje.html', mensaje=mensaje)
          
@app.route('/resultado')
def resultado():
    # .... inicio deteccion ....................
    
    global probabilidad
    global etiqueta
    
    # Creando una clase para la configuracin de la prediccin:
    class ConfiguracionesHojas(Config): 
        NAME = "configuraciones_hojas"
        NUM_CLASSES = 1 + 2  # define o nmero de classes (background + HojaTomate + NoHojaTomate)
        #IMAGE_MIN_DIM = 227
        #IMAGE_MAX_DIM = 512
        GPU_COUNT = 1
        IMAGES_PER_GPU = 1
        DETECTION_MAX_INSTANCES = 1

    # Creando una funcion que muesstra n imagenes de test, cada una con su mscara real e com la caja prevista por el modelo:
    def detectar_objeto(imagen, modelo, cfg, nombre_archivo):
        lista_clase_deteccion = ['BG','HojaTomate','NoHojaTomate']   # clases para hacer deteccin
        global datos
        imagen_centrada = mold_image(imagen, cfg) # aplicando centering
        muestra = expand_dims(imagen_centrada, 0) # creando la dimensin extra de batch_size
        datos = modelo.detect(muestra, verbose=0)[0] # haciendo la prediccin de la imagen
        
        plt.imshow(imagen, aspect = 'auto') # muestra la imagen
        
        ejes = plt.gca()
        global probabilidad
        global etiqueta
        # plotando las cajas previstas:
        for caja in datos['rois']:
            y1, x1, y2, x2 = caja
            ancho, altura = x2 - x1, y2 - y1
            rectangulo = Rectangle((x1, y1), ancho, altura, fill=True, alpha=0.3, color='red')
            ejes.add_patch(rectangulo) # adiciona el rectngulo previsto por el modelo en la figura
            probabilidad =float(datos['scores'])
            indice_clase = int(datos['class_ids'])
            etiqueta = lista_clase_deteccion[indice_clase]
            ejes.text(x1+5, y1+20, lista_clase_deteccion[indice_clase] +': ' + str(round(probabilidad*100,1))+'%')
            plt.axis('off')      # deshabilita los ejes de la figura
            plt.savefig('./static/output/'+nombre_archivo, bbox_inches='tight',pad_inches = 0) 
        
        plt.show()

        print('probabilidad de deteccion =',probabilidad,' etiqueta =',etiqueta)
        
    # ... recupera imagen de la carpeta de archivos subidos ..
    for nombre_archivo in os.listdir('./static/upload'):
        print('archivo: ',nombre_archivo)
        
    # Cargando la imagen:
    imagen = imread('./static/upload/'+nombre_archivo) 

    cfg = ConfiguracionesHojas()

    modelo = MaskRCNN(mode='inference', model_dir='C:/Users/HP/UNLP/modelo_testando', config=cfg)
    modelo.load_weights('C:/Users/HP/UNLP/modelo/modelo_deteccion/mask_rcnn_Detecta.h5', by_name=True)

    # Mostrando la prediccin de la imagen:
    detectar_objeto(imagen, modelo, cfg, nombre_archivo)

    # .... fin deteccion .......................
        
    probabilidadPrecision = 0.70             # ...valor: probabilidadPrecision ...
    
    print('probabilidad deteccion =', probabilidad, 'etiqueta =', etiqueta)
    
    if probabilidad<probabilidadPrecision or etiqueta!='HojaTomate':    # ... controla que la imagen subida corresponda a la clase definidad para la deteccion ...   
        mensaje='<<< ... ERROR ... >>> La imagen subida no corresponde a <<< ...'+ etiquetas[0] +'... >>> o tiene una probabilidad inferior a '+str(probabilidadPrecision*100)+'%, suba otra imagen.'
        return render_template('mensaje2.html', mensaje=mensaje, nombre_archivo=nombre_archivo)
    
    
    # .... inicio clasificacion .......................
    
    probabilidad_clasificacion = -1   # iniciando variable probabilidad de clasifcacion ...
        
    # ... recupera imagen de la carpeta de archivos subidos ..
    for nombre_archivo in os.listdir('./static/upload'):
        print('archivo: ',nombre_archivo)
        
    # ...recortando el area del objeto detectado ...
    nombre_area_detectada = 'Area' + nombre_archivo
    img = cv2.imread('./static/upload/'+nombre_archivo,cv2.IMREAD_UNCHANGED) 
    
    x = datos['rois'][0][0]
    y = datos['rois'][0][1]
    w = datos['rois'][0][2]
    h = datos['rois'][0][3]
    crop_img = img[x:w, y:h]
    cv2.imwrite('./static/output/'+nombre_area_detectada, crop_img)
    
    names = ['TOMATE MANCHA BACTERIANA','TOMATE SANO','TOMATE TIZON TEMPRAMO']
    modelt = load_model("C:/Users/HP/UNLP/modelo/modelo_clasificacion/model_VGG16.h5")

    img = cv2.imread('./static/output/Area'+nombre_archivo) 
    
    width = 224    # int(img.shape[1])
    height = 224   # int(img.shape[0])
    dim = (width, height)
    # resize image
    imaget = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

    xt = np.asarray(imaget)
    xt=preprocess_input(xt)
    xt = np.expand_dims(xt,axis=0)
    preds = modelt.predict(xt)

    print('preds[0]:',preds[0])   # ... tiene las probabilidades por cada clase ...
    print(names[np.argmax(preds)],preds[0][np.argmax(preds)]*100.0,'%')
    plt.imshow(cv2.cvtColor(np.asarray(imaget),cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()    
    
    probabilidad_clasificacion = preds[0][np.argmax(preds)]  # iniciando variable probabilidad de clasifcacion ...
    
    
        
    if probabilidad_clasificacion== -1:    # ... controla que la imagen subida corresponda a la clase definidad para la deteccion ...   
        mensaje='<<< ... ERROR ... >>> La imagen subida no corresponde a <<< ...'+ etiquetas[0] +'... >>> o tiene una probabilidad inferior a '+str(probabilidadPrecision*100)+'%, que no corresponde a las enfermedades definidas, suba otra imagen.'
        return render_template('mensaje2.html', mensaje=mensaje, nombre_archivo=nombre_archivo)
  


    # inicializando variables para prxima consulta ..
    probabilidad = -1.00  # si la probabilidad es menor a la del umbral definido, se queda con el valor inicial negativo
    etiqueta=''     # guarda clase detectada ...
  
    
    
    answer = np.argmax(preds[0])
    
    sintEnfermedad=''       # ... sintoma enfermedad ...
    prevEnfermedad=''       # ... prevencion enfermedad ...
    tratEnfermedad=''       # ... tratamiento enfermedad ...
    fotoEnfermedad=''
    respuesta=''
    if answer == 0:
        print("prediccion: Mancha Bacteriana")
        print('Probabilidad clasificacion=',preds[0][np.argmax(preds)]*100.0,'%')
        probabilidad_clasifcacion= round(preds[0][np.argmax(preds)]*100.0,1)
        respuesta='Mancha Bacteriana'
        fotoEnfermedad='fotoManchaBacteriana.jpg'
        sintEnfermedad = './static/img/enfermedades/sintManchaBacteriana.html'    # ... sintoma enfermedad ...
        prevEnfermedad = './static/img/enfermedades/prevManchaBacteriana.html'    # ... prevenir enfermedad ...
        tratEnfermedad = './static/img/enfermedades/tratManchaBacteriana.html'    # ... tratamiento enfermedad ...

    elif answer == 1:
        print("prediccion: Tomate Sano")
        print('Probabilidad clasificacion=',preds[0][np.argmax(preds)]*100.0,'%')
        probabilidad_clasifcacion= round(preds[0][np.argmax(preds)]*100.0,1)
        respuesta='Tomate Sano'
        fotoEnfermedad='fotoTomateSano.jpg'
        sintEnfermedad=''       # ... sintoma enfermedad ...
        prevEnfermedad=''       # ... prevencion enfermedad ...
        tratEnfermedad=''       # ... tratamiento enfermedad ...
    else:                  # ...  answer == 2
        print("prediccion: Tizon Temprano")
        print('Probabilidad clasificacion=',preds[0][np.argmax(preds)]*100.0,'%')
        probabilidad_clasifcacion= round(preds[0][np.argmax(preds)]*100.0,1)
        respuesta='Tizon Temprano'
        fotoEnfermedad='fotoTizonTemprano.jpg'
        sintEnfermedad = './static/img/enfermedades/sintTizonTemprano.html'    # ... sintoma enfermedad ...
        prevEnfermedad = './static/img/enfermedades/prevTizonTemprano.html'    # ... prevenir enfermedad ...
        tratEnfermedad = './static/img/enfermedades/tratTizonTemprano.html'    # ... tratamiento enfermedad ...
    
    mensaje = respuesta  
    return render_template('prediccion.html',mensaje=mensaje,probabilidad=probabilidad_clasifcacion,nombre_archivo=nombre_archivo,fotoEnfermedad=fotoEnfermedad,sintEnfermedad=sintEnfermedad,prevEnfermedad=prevEnfermedad,tratEnfermedad =tratEnfermedad)
    # .... fin clasificacion .......................  
    
    
@app.route('/credito')
def credito():
    return render_template('credito.html')


@app.errorhandler(404)
def page_not_found(error):
    return "<h1>Esta pagina no existe ...</h1>",404 


if __name__ == '__main__':
    app.run(debug=False,host="0.0.0.0")
```
